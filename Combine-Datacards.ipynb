```python
norm_to_1fb = False
# norm_to_1fb = True

run_dissociative_signal = True
# run_dissociative_signal = False

# upper_mass_ratio_cut = True
upper_mass_ratio_cut = False

# alpha = 'aCw'
alpha = 'a0w'

# categories = [ 'muon-2017', 'electron-2017', 'muon-2018', 'electron-2018' ]
categories = [ 'muon-2018', 'electron-2018' ]
```

```python
df_sum_weights = {}
file_str_ = "datacards_norm-F_diss-T_mass-F_lim-nom/table_systematics"
if norm_to_1fb:
    file_str_ += "-norm_to_1fb"
if upper_mass_ratio_cut:
    file_str_ += "-mass_cut"
if run_dissociative_signal:
    file_str_ += "-diss"
for category__ in categories:
    file_path_ = ""
    if category__ == 'muon-2017':
        file_path_ = file_str_ + "-2017-muon" + ".h5"
    elif category__ == 'electron-2017':
        file_path_ = file_str_ + "-2017-electron" + ".h5"
    elif category__ == 'muon-2018':
        file_path_ = file_str_ + "-"+alpha+"-2018-muon" + ".h5"
    elif category__ == 'electron-2018':
        file_path_ = file_str_ + "-"+alpha+"-2018-electron" + ".h5"
       
    print( file_path_ )
    with pd.HDFStore( file_path_, 'r' ) as store_:
        print ( list( store_ ) )
        df_sum_weights[ category__ ] = store_[ "table_systematics" ]
```

```python
df_sum_weights_sup = {}
file_str_ = "datacards_norm-F_diss-T_mass-F_lim-sup/table_systematics"
if norm_to_1fb:
    file_str_ += "-norm_to_1fb"
if upper_mass_ratio_cut:
    file_str_ += "-mass_cut"
if run_dissociative_signal:
    file_str_ += "-diss"
for category__ in categories:
    file_path_ = ""
    if category__ == 'muon-2017':
        file_path_ = file_str_ + "-2017-muon" + ".h5"
    elif category__ == 'electron-2017':
        file_path_ = file_str_ + "-2017-electron" + ".h5"
    elif category__ == 'muon-2018':
        file_path_ = file_str_ + "-"+alpha+"-2018-muon" + ".h5"
    elif category__ == 'electron-2018':
        file_path_ = file_str_ + "-"+alpha+"-2018-electron" + ".h5"
       
    print( file_path_ )
    with pd.HDFStore( file_path_, 'r' ) as store_:
        print ( list( store_ ) )
        df_sum_weights_sup[ category__ ] = store_[ "table_systematics" ]
```

```python
df_sum_weights_inf = {}
file_str_ = "datacards_norm-F_diss-T_mass-F_lim-inf/table_systematics"
if norm_to_1fb:
    file_str_ += "-norm_to_1fb"
if upper_mass_ratio_cut:
    file_str_ += "-mass_cut"
if run_dissociative_signal:
    file_str_ += "-diss"
for category__ in categories:
    file_path_ = ""
    if category__ == 'muon-2017':
        file_path_ = file_str_ + "-2017-muon" + ".h5"
    elif category__ == 'electron-2017':
        file_path_ = file_str_ + "-2017-electron" + ".h5"
    elif category__ == 'muon-2018':
        file_path_ = file_str_ + "-"+alpha+"-2018-muon" + ".h5"
    elif category__ == 'electron-2018':
        file_path_ = file_str_ + "-"+alpha+"-2018-electron" + ".h5"
       
    print( file_path_ )
    with pd.HDFStore( file_path_, 'r' ) as store_:
        print ( list( store_ ) )
        df_sum_weights_inf[ category__ ] = store_[ "table_systematics" ]
```

```python
df_uncertainties_relative = {}
for category__ in categories:
    if run_dissociative_signal:
        df_up = df_sum_weights_sup[ category__ ].iloc[ 0, : ] - df_sum_weights[ category__ ].iloc[ 0, : ]
        df_up['Background'] = -1.0 
        df_dw = df_sum_weights_inf[ category__ ].iloc[ 0, : ] - df_sum_weights[ category__ ].iloc[ 0, : ] 
        df_dw['Background'] = -1.0
        df1 = pd.DataFrame(df_up) 
        df1 = df1.transpose()
        df1 = df1.rename(index={'central': 'syst_diss_up'})
        df2 = pd.DataFrame(df_dw) 
        df2 = df2.transpose()
        df2 = df2.rename(index={'central': 'syst_diss_dw'})
        df3 = pd.concat([df1, df2], ignore_index = False) 
        display(df3)
        df_sum_weights[ category__ ] = pd.concat([df_sum_weights[ category__ ], df3], ignore_index = False) 
    df__ = df_sum_weights[ category__ ]
    df_uncertainties_relative[ category__ ] = df__.iloc[ 1:, : ].where( ~( df__.iloc[ 1:, : ] != -1. ) , df__.iloc[ 1:, : ].divide( df__.iloc[ 0 ] ) )
       
```
